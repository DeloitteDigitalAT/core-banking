```python
fake = Faker() 
def create_data(x): 
  
    # dictionary 
    client ={} 
    for i in range(0, x): 
        client[i]={} 
        client[i]['socialsecuritynumber']= fake.ssn() 
        client[i]['firstname']= faker.first_name() 
        client[i]['lastname']= faker.last_name() 
        client[i]['id']= faker.pyint(min_value=10000, max_value=99999, step=1)
        client[i]['sex']= faker.random_element(elements=('M', 'F')) 
        client[i]['dateofbirth']= faker.date() 
        client[i]['createdDate'] = datetime.datetime.now()
        client[i]['citizenship'] = faker.random_element(elements=('Austria', 'Germany','Tuvalu','India','Armenia','USA','UK','Portugal'))
        client[i]['flagged_customer'] = boolean(chance_of_getting_true=8)

    return client
```

```python
def create_accounts(x): 
  
    # dictionary 
    accounts ={} 
    for i in range(0, x): 
        accounts[i]={} 
        accounts[i]['clientid']= faker.random_element(elements = Client['clientid']) 
        accounts[i]['account_id']= faker.bban()
        accounts[i]['accountdate']= faker.date() 
        accounts[i]['createdDate'] = datetime.datetime.now()
        accounts[i]['frequency']= faker.random_element(elements=('Monthly', 'Quaterly','Per Transaction')) 
        

    return accounts
```

```python
def create_transactions(x): 
  
    # dictionary 
    transactions ={} 
    for i in range(0, x): 
        transactions[i]={} 
        transactions[i]['accountid']= faker.random_element(elements = Account['account_id'])  
        transactions[i]['amount']= faker.pydecimal(min_value=0, max_value=100000, right_digits=2)
        transactions[i]['accountIban'] = faker.iban()
        transactions[i]['balance']= faker.pydecimal(min_value=0, max_value=100000, right_digits=2)
        transactions[i]['transactiondate']= faker.date() 
        transactions[i]['type']= faker.random_element(elements=('Debit', 'Credit'))
        transactions[i]['createdDate'] = datetime.datetime.now()
       
        

    return transactions
```

```python
def create_orders(x): 
  
    # dictionary 
    orders ={} 
    for i in range(0, x): 
        orders[i]={}  
        orders[i]['orderid']= faker.pyint(min_value=0, max_value=10000000, step=1)
        orders[i]['bankto']= fake.swift()
        orders[i]['accountto']= faker.bban()
        orders[i]['referance']= faker.text(max_nb_chars = 15)
        orders[i]['createdDate'] = datetime.datetime.now()
        

    return orders
```

```python
def create_overdraft(x): 
  
    # dictionary 
    overdraft ={} 
    for i in range(0, x): 
        overdraft[i]={} 
        overdraft[i]['accountid']= faker.random_element(elements = Account['account_id'])  
        overdraft[i]['amount']= faker.pydecimal(min_value=0, max_value=100000, right_digits=2)
        overdraft[i]['balance']= faker.pydecimal(min_value=0, max_value=100000, right_digits=2)
        overdraft[i]['defaultcount']= faker.pyint(min_value=0, max_value=5, step=1) 
        overdraft[i]['createdDate'] = datetime.datetime.now()
        

    return overdraft
```

```python
def create_addressdetails(x): 
  
    # dictionary 
    address ={} 
    for i in range(0, x): 
        address[i]={} 
        address[i]['clientid']= faker.random_element(elements = Client['clientid']) 
        address[i]['Houseno']= faker.building_number()
        address[i]['City']= faker.city() 
        address[i]['Pincode']= faker.random_element(elements=('1010','1200','2020','2010','3010','4010')) 
        address[i]['Streetname']= faker.street_address()
        address[i]['createdDate'] = datetime.datetime.now()
        

    return address
```

```python
def create_Contact_Details(x): 
  
    # dictionary 
    contacts ={} 
    for i in range(0, x): 
        contacts[i]={} 
        contacts[i]['clientid']= faker.random_element(elements = Client['clientid']) 
        contacts[i]['Phonenumber']= faker.msisdn()
        contacts[i]['createdDate'] = datetime.datetime.now()
        #friends_data[i]['Email_id']=  Client['last_name']+'.'+Client['first_name']+'@gmail.com' # bring through join
    return contacts
```

```python
def create_Indicators(x): 
  
    # dictionary 
    indicators ={} 
    for i in range(0, x): 
        indicators[i]={} 
        indicators[i]['clientid']= faker.random_element(elements = Client['clientid']) 
        indicators[i]['EmpStatus']= faker.random_element(elements = ('Blue Collar','White Collar','Part time','Unemployed'))
        indicators[i]['IncomeGroup']= faker.random_element(elements = ('<10000','10000-20000','20000-30000','30000-40000','40000-50000','60000<'))
        indicators[i]['IsRiskGroup']= faker.boolean()
        indicators[i]['IsPEP']= faker.boolean()
        indicators[i]['marital_Status']= faker.random_element(elements = ('Married','Single','Divorced','Engaged','Other'))
        indicators[i]['createdDate'] = datetime.datetime.now()
    return indicators
```
